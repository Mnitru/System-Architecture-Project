```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file Excel
df = pd.read_excel("data.xlsx")

# Sắp xếp theo source (nếu cần)
df = df.sort_values(by="source")

# Vẽ biểu đồ
plt.figure()
plt.bar(df["source"], df["time"])

# Label
plt.xlabel("Source")
plt.ylabel("Time")
plt.title("Source vs Time")

# Hiển thị
plt.show()
```

```python
plt.figure()
plt.plot(df["source"], df["time"], marker='o')

plt.xlabel("Source")
plt.ylabel("Time")
plt.title("Source vs Time (Line Chart)")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data.xlsx")

plt.figure()

# Vẽ bar chart
plt.bar(df["source"], df["time"])

# Set trục X = 1, 2, 3
plt.xticks([1, 2, 3])

# Set trục Y từ 0 -> 120000
plt.ylim(0, 120000)

# Label
plt.xlabel("Source")
plt.ylabel("Time (ms)")
plt.title("Source vs Time")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data.xlsx")

plt.figure()

# Vẽ line chart
plt.plot(df["source"], df["time"], marker='o')

# Trục X: chỉ hiện 1, 2, 3
plt.xticks([1, 2, 3])

# Trục Y: từ 0 đến 120000 ms
plt.ylim(0, 120000)

# Chia mốc trục Y cho dễ nhìn
plt.yticks(range(0, 120001, 20000))

# Label
plt.xlabel("Source")
plt.ylabel("Time (ms)")
plt.title("Source vs Time (Line Chart)")

plt.grid(True)

plt.show()
```
